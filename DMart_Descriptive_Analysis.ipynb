## Task 5: Grouped Analysis

```python
# What is the distribution of customers by age and gender? Are certain age groups more active in purchasing?

sns.displot(df, x='Customer Age', hue='Gender', kind='kde')
plt.title("Customer Age Distribution by Gender")
plt.show()
```

```python
# How does customer subscription status affect total purchase value?

subscription_data = df.groupby('Subscription')['Total Order Value'].sum()
df['Total Order Value'] = pd.to_numeric(df['Total Order Value'], errors='coerce')
plt.figure(figsize=(20,5)) 
sns.barplot(hue=subscription_data.index, y=subscription_data.values, palette="viridis", legend=True)
plt.xlabel("Subscription Type")
plt.ylabel("Total Order Value")
plt.title("Total Order Value by Subscription Type")
plt.show()
```

```python
# How does customer subscription status affect total purchase value by Category?

sub_category = df.groupby('Subscription')['Category'].value_counts().reset_index(name='Count')
plt.figure(figsize=(20,5)) 

sns.barplot(x='Subscription', y='Count', hue='Category', data=sub_category, palette="viridis")
plt.xlabel("Subscription Type")
plt.ylabel("Total Order Value")
plt.title("Total Order Value by Subscription Type")
```

```python
# What are the top-selling products based on total order value?

df.groupby('Product Name')['Total Order Value'].sum()

# Harpic Bathroom Cleaner - Floral is selling at top with revenue of 828936.33 highest among all products
```

## Task 6: Correlation Heatmap

```python
df.corr(numeric_only=True)
```

```python
corr_matrix = df[['MRP', 'Discount Price', 'Customer Age', 'Total Order Value']].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, vmax=1, vmin=-1, cmap='viridis', linewidths=0.5, fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```
